# Gradient Based Learning

```python
import numpy as np
import matplotlib.pyplot as plt

function_x = np.linspace(-3, 3, 100)
function_y = function_x**2

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(function_x, function_y)

x = 2
y = x**2
lr = 0.1   # learning rate
ax.scatter(x, y, color='red', s=300)     # 초기값

for _ in range(10):
    dy_dx = 2*x       # x**2의 미분
    x = x - lr*dy_dx  # learning rate만큼 이동
    y = x**2
    
    ax.scatter(x, y, color='red', s=100)   # s: size
```

```python
import numpy as np
import matplotlib.pyplot as plt

function_x = np.linspace(-3, 3, 100)
function_y = 2*function_x**2

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(function_x, function_y)

x = 3
y = 2*x**2
lr = 0.4   # learning rate
ax.scatter(x, y, color='red', s=300)     # 초기값

for _ in range(10):
    dy_dx = 4*x       # x**2의 미분
    x = x - lr*dy_dx  # learning rate만큼 이동
    y = 2*x**2
    
    ax.scatter(x, y, color='red', s=100)   # s: size
```

```python
import numpy as np
import matplotlib.pyplot as plt

function_x = np.linspace(-3, 3, 100)
function_y = function_x**3

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(function_x, function_y)    # 사실은 이 선을 모르는 상황인 것

x = 3
y = x**3
lr = 0.01   # learning rate
ax.scatter(x, y, color='red', s=300)     # 초기값

for _ in range(30):
    dy_dx = 3*x**2       # x**2의 미분
    x = x - lr*dy_dx  # learning rate만큼 이동
    y = x**3
    
    ax.scatter(x, y, color='red', s=100)   # s: size
```
